# Batch inversion
Batch inversion is the fact to allow to invert multiple surveys in once. Here we reuse a lot of stuff from the time-lapse implementation.

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import os
import sys
sys.path.append((os.path.relpath('../..'))) # add here the relative path of the API folder

from api.R2 import R2
```

```python
k = R2()
k.createBatchSurvey('../../api/test/testTimelapse')
```

And inversion as usual.

```python
k.invert(iplot=False)
```

And each inverted section can be shown using the same function `R2.showResults()`.

```python
k.showResults(index=0, contour=True, vmin=1.5, vmax=2)
k.showResults(index=1, contour=True, vmin=1.5, vmax=2)
k.showResults(index=2, contour=True, vmin=1.5, vmax=2)
k.showResults(index=3, contour=True, vmin=1.5, vmax=2)
```
